```python
import pandas as pd

# make sure date is in datetime format
df['date'] = pd.to_datetime(df['date'])

# sort by date = df.sort_values('date')
commodity_prices = df.sort_values('date')

# Month over Month growth % for wheat price
df['wheat_mom_growth_pct'] = df['wheat_usd_mt'].pct_change() * 100

# Year over Year growth % for wheat price
df['wheat_yoy_growth_pct'] = df['wheat_usd_mt'].pct_change(12) * 100

print(df[['date', 'wheat_usd_mt', 'wheat_mom_growth_pct', 'wheat_yoy_growth_pct']].head(780))
```

```python
df['date'] = pd.to_datetime(df['date'])
commodity_prices = df.sort_values('date')
df['maize_mom_growth_pct'] = df['maize_usd_mt'].pct_change() * 100
df['maize_yoy_growth_pct'] = df['maize_usd_mt'].pct_change(12) * 100
print(df[['date', 'maize_usd_mt', 'maize_mom_growth_pct', 'maize_yoy_growth_pct']].head(780))
```

```python
df['date'] = pd.to_datetime(df['date'])
commodity_prices = df.sort_values('date')
df['soybean_oil_mom_growth_pct'] = df['soybean_oil_usd_mt'].pct_change() * 100
df['soybean_oil_yoy_growth_pct'] = df['soybean_oil_usd_mt'].pct_change(12) * 100
print(df[['date', 'soybean_oil_usd_mt', 'soybean_oil_mom_growth_pct', 'soybean_oil_yoy_growth_pct']].head(780))
```

```python
df['wheat_mom_growth_pct'] = df['wheat_mom_growth_pct'].fillna('0')
df['wheat_yoy_growth_pct'] = df['wheat_yoy_growth_pct'].fillna('0')
df['maize_mom_growth_pct'] = df['maize_mom_growth_pct'].fillna('0')
df['maize_yoy_growth_pct'] = df['maize_yoy_growth_pct'].fillna('0')
df['soybean_oil_mom_growth_pct'] = df['soybean_oil_mom_growth_pct'].fillna('0')
df['soybean_oil_yoy_growth_pct'] = df['soybean_oil_yoy_growth_pct'].fillna('0')
```

```python
df.to_csv("commodity_prices.csv", index=False)
```
